```python
# Drop duplicate columns

duplicate_cols = [col for col in df.columns if col.endswith('.1')]
df = df.drop(columns=duplicate_cols)

df.shape
```

```python
# Drop columns where more than 90% of values are unique strings

# Exclude date columns
date_cols = ["ListingContractDate", "CloseDate", "PurchaseContractDate", "ContractStatusChangeDate"]

high_cardinality = [
    col for col in df.columns
    if col not in date_cols and df[col].nunique() / len(df) > 0.9
]

print(f"Suggested columns to drop: {high_cardinality}")

df = df.drop(columns=high_cardinality)

df.shape
```

```python
# Drop columns that have more than 50% NaN

cols_more_than_50_nan = [
    col for col in df.columns
    if col not in date_cols and df[col].isna().mean() > 0.5
]

print("Columns with more than 50% missing values:")
print(cols_more_than_50_nan)

# Drop those columns
df = df.drop(columns=cols_more_than_50_nan)

df.shape
```

Market Analysis Fields = columns that help you analyze the real estate market

Metadata Fields = columns that mainly describe the record, system, import process, or listing identity, but do not help much with market insights

```python
# Identify Market Analysis Fields

analysis_fields = ['OriginalListPrice', 'LivingArea', 'ListPrice', 'DaysOnMarket', 'AttachedGarageYN', 'ParkingTotal', 'LotSizeAcres', 'YearBuilt', 'BathroomsTotalInteger', 'BedroomsTotal', 'FireplaceYN', 'Stories', 'Levels', 'LotSizeArea', 'MainLevelBedrooms', 'NewConstructionYN', 'GarageSpaces','LotSizeSquareFeet']

target_col = []
```

```python
# Change YN columns to 0/1 Binary variables, with NaN = 0

YN_columns = ['NewConstructionYN', 'FireplaceYN', 'AttachedGarageYN']

for col in YN_columns:
    df[col] = (
        df[col]
        .replace({"Yes": 1, "No": 0})
        .fillna(0)
        .astype(int)
    )

df.head()
```

```python
# Create year/month features from date columns

df['ListingContractDate'] = pd.to_datetime(df['ListingContractDate'], errors='coerce')
df['list_year'] = df['ListingContractDate'].dt.year
df['list_month'] = df['ListingContractDate'].dt.month
df['list_yrmo'] = df['list_year'] * 100 + df['list_month']

print("Listings date range:", df['ListingContractDate'].min().date(), "to", df['ListingContractDate'].max().date())
```
